```python
from tokenizer import Tokenizer
```

```python
with open("Goethe--Faust.txt", "r") as f:
    content = f.read()

t = Tokenizer()
tokens = t.tokenize(content, 512)
t.print_mapping_counts()
```

```python
import seaborn as sns
import numpy as np

stats = np.array(t.stats)
statsT = stats.T

# Absolute
sns.lineplot(x=statsT[1], y=statsT[0])
```

```python
# Percent
sns.lineplot(x=statsT[1], y=(statsT[0]/statsT[0][0]))
```
